```python
import pandas as pd
import json
import requests
import folium
import urllib
#opgave 3
monumenter = pd.read_csv("https://wfs-kbhkort.kk.dk/k101/ows?service=WFS&version=1.0.0&request=GetFeature&typeName=k101:monumenter&outputFormat=csv&SRSNAME=EPSG:4326")
def monumentById(monumentId = None, monumentName = None):
    if(monumentId != None):
        mon_Id = monumenter.loc[monumenter['id']==monumentId]
        
        longitude = mon_Id.iloc[0,7]
        latitude = mon_Id.iloc[0,8]
        return( latitude, longitude) 
    else:
        mon_Name = monumenter.loc[monumenter['navn']==monumentName]
        longitude = mon_Name.iloc[0,7]
        latitude = mon_Name.iloc[0,8]
        return( latitude, longitude) 

position = monumentById(monumentId=50063)
print(monumentById(monumentId=50063))    
    
#opgave 3a   
map_osm = folium.Map(location=(position), zoom_start=10) 
folium.Marker(position).add_to(map_osm)      
map_osm.save('./osm.html')
map_osm

#opgave 5


map_osm1 = folium.Map(location=(position), zoom_start=10)
for x in range(monumenter.shape[0]):
    folium.Marker((monumenter.iloc[x,8], monumenter.iloc[x,7])).add_to(map_osm1)
map_osm1.save('./osm.html')
map_osm1
```
